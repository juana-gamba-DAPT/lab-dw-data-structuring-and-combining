Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Crear pivot table para total revenue por canal de ventas
pivot_revenue = pd.pivot_table(
    pdi,
    values='total_claim_amount',
    index='sales_channel',
    aggfunc='sum'
).round(2)

# Ordenar de mayor a menor
pivot_revenue = pivot_revenue.sort_values(by='total_claim_amount', ascending=False)

print(pivot_revenue)
```

```python
#2. Create a pivot table that shows the average customer lifetime value per gender and education
#  level. Analyze the resulting table to draw insights.
pivot_value=pd.pivot_table(
    pdi,
    values='customer_lifetime_value',
    index={'gender','education'},
    aggfunc='mean'
).round(0).sort_values(by='customer_lifetime_value',ascending=False)
print(pivot_value)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

```python
# Asegro de que 'month' esté en formato texto si no lo está
pdi['month'] = pdi['month'].astype(str)

# Agrupamos por mes y tipo de póliza, y contamos las quejas
complaints_summary = (
    pdi.groupby(['policy_type', 'month'])['number_of_open_complaints']
    .count()
    .reset_index()
    .rename(columns={'number_of_open_complaints': 'number_of_complaints'})
)
```

```python
# Ordenamos por número de quejas descendente
complaints_summary = complaints_summary.sort_values(by='number_of_complaints', ascending=False)
```

```python
print(complaints_summary)
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*
